```python
import numpy as np
import pandas as pd

n = 50000

# Listas base
nombres = ["Juan", "Ana", "Luis", "María", "Carlos", "Sofía", "Pedro", "Lucía"]
apellidos = ["Pérez", "Gómez", "López", "Rodríguez", "Hernández", "Martínez"]
paises = ["Guatemala", "México", "Colombia", "Perú", "Chile", "Argentina"]

# Crear datos
Datos = {
    "id": np.arange(1, n+1),
    "nombre": np.random.choice(nombres, n),
    "apellido": np.random.choice(apellidos, n),
    "Genero": np.random.choice(["F", "M"], n),
    "Pais": np.random.choice(paises, n),
    "estado": np.random.choice(["soltero(a)", "casado(a)"], n),
    "salario": np.random.randint(1500, 10001, n),
    "Edad": np.random.randint(18, 81, n),
    "registro": pd.to_datetime(
        np.random.choice(pd.date_range("2026-01-01", "2026-12-31"), n)
    )
}

DatosDF = pd.DataFrame(Datos)
```

```python
# Desviación estándar
print(DatosDF["salario"].std())
print(DatosDF["Edad"].std())
```
